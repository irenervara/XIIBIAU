```python
import pandas as pd
import numpy as np
import geopandas as gpd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('XII-BIAU-Panorama_Obras-listado-prueba.xlsx') # dataset obras
df_places = pd.read_csv('df_ambitos.csv', encoding='utf-8') # relación país-ámbito
```

```python
df.head()
```

```python
df.columns
```

```python
# SELECCIÓN DE COLUMNAS

# rename
rename_cols = {'Tipo_Obra':'tipo_obra',
               'dimension':'escala',
               'sociedad':'im_sociedad',
               'economa':'im_econom',
               'medio_ambiente':'im_ma',
               'transversal':'im_trans',
               'pais_de_emplazamiento_de_la_obra':'pais',
               'acronimo':'acr',
               'fecha_fin':'año',
               'imagen1baja_URL':'img_url',
               'FICHA_URL':'web_url'}

# selección + rename
dff = df[['Tipo_Obra',
          'dimension',
          'usos',
          'sociedad',
          'economa',
          'medio_ambiente',
          'transversal',
          'titulo_Obra', 
          'acronimo',
          'fecha_fin', 
          'pais_de_emplazamiento_de_la_obra', 
          'presupuesto',
          'moneda',
          'autores',
          'imagen1baja_URL',
          'FICHA_URL']].rename(columns=rename_cols)
```

```python
dff
```

```python
# AÑO

for i in ['año']:
    dff[i] = dff[i].str.replace('No procede', '')
    dff[i] = dff[i].str[-4:]
    dff[i] = dff[i].replace(r'^\s*$', np.NaN, regex=True) # replace empty string with NaN
    dff[i] = dff[i].fillna('Sin datos')
```

```python
dff.año
```

```python
# DEFINIR TRAMOS "ESCALA"

# pendiente de acabar
def func_escala(row):
    if row['escala']<=250:
        return 's'
    elif row['escala']<=500:
        return 'm'
    elif row['escala']<=1000:
        return 'l'
    else:
        return 'xl'

# pendiente de acabar
dff['escala_tramos'] = dff.apply(lambda row: func_escala(row), axis=1)
```

```python
# MONEDA

dict_moneda = {'Dólares':'$', 'Euro':'€'}
dff = dff.replace({'moneda':dict_moneda})
```

```python
# DEFINIR ÁMBITOS

dict_ambitos = dict(zip(df_places['País'], df_places['Ámbito']))
dff.loc[dff['pais'].isin(dict_ambitos.keys()), 'ambito'] = dff['pais'].map(dict_ambitos)
```

```python
dff = dff[['tipo_obra', 'escala', 'escala_tramos', 'usos', 'im_sociedad', 'im_econom', 'im_ma','im_trans', 'titulo_Obra', 'acr', 'año', 'pais', 'ambito', 'presupuesto', 'moneda','autores', 'img_url', 'web_url']]
```

```python
dff['categoria'] = ['obras', 'obras', 'trayectoria', 'publicacion', 'prog_doc', 'acc']
dff['city'] = ['caracas', 'medellin', 'antioquia', 'santiago de chile', 'iquique', 'la paz']
dff['lat'] = ['10.500000', '6.230833', '7.883609', '-34.672501', '-20.216700', '-16.489689']
dff['lon'] = ['-66.916664', '-75.590553', '-76.632111', '-58.449722', '-70.142220', '-68.119293']
```
