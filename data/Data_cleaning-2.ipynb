## OBRAS

```python
df_obras = pd.read_csv('recibido/ivnf4_chronoforms_data_xii-biau-panorama-de-obras-final.csv')
df_obras.head()
```

```python
df_obras.columns
```

```python
# AÑADIR URL IMG
suffix_obras = 'http://www.bienalesdearquitectura.es/images/XII_BIAU/Panorama_obras/'
df_obras['img_url'] = suffix_obras+df_obras['imagen1baja']

# AÑADIR URL WEB
suffix_url_web = 'http://www.bienalesdearquitectura.es/index.php/es/?option=com_content&view=article&id='
df_obras['ficha_url'] = df_obras['acronimo']
```

```python
# FECHA FIN: UNIFICAR LAS 2 COLUMNAS

# pendiente de acabar
def func_fecha(row):
    if row['fecha_fin']=='No procede':
        return row['fecha_fin_espana']
    else:
        return row['fecha_fin']

# pendiente de acabar
df_obras['fecha'] = df_obras.apply(lambda row: func_fecha(row), axis=1)
```

```python
# SELECCIÓN DE COLUMNAS

# rename
rename_cols = {'Tipo_Obra':'tipo_obra',
               'dimension':'escala',
               'sociedad':'im_sociedad',
               'economa':'im_econom',
               'medio_ambiente':'im_ma',
               'transversal':'im_trans',
               'pais_de_emplazamiento_de_la_obra':'pais',
               'situacion_obra':'direccion',
               'acronimo':'acr',
               'fecha':'año'}

# selección + rename
df_obras = df_obras[['Tipo_Obra',
          'dimension',
          'usos',
          'sociedad',
          'economa',
          'medio_ambiente',
          'transversal',
          'titulo_Obra', 
          'acronimo',
          'fecha', 
          'situacion_obra',
          'pais_de_emplazamiento_de_la_obra', 
          'presupuesto',
          'moneda',
          'autores',
          'img_url',
          'ficha_url']].rename(columns=rename_cols)
```

```python
# AÑO

for i in ['año']:
    df_obras[i] = df_obras[i].str.replace('No procede', '')
    df_obras[i] = df_obras[i].str[-4:]
    df_obras[i] = df_obras[i].replace(r'^\s*$', np.NaN, regex=True) # replace empty string with NaN
    df_obras[i] = df_obras[i].fillna('Sin datos')
```

```python
# DEFINIR TRAMOS "ESCALA"

# convertir 'escala' a numeric
df_obras['escala'] = df_obras['escala'].str.replace(',', '.')
df_obras['escala'] = df_obras['escala'].astype('float')

# pendiente de acabar
def func_escala(row):
    if row['escala']<=250:
        return 's'
    elif row['escala']<=500:
        return 'm'
    elif row['escala']<=1000:
        return 'l'
    else:
        return 'xl'

# pendiente de acabar
df_obras['escala_tramos'] = df_obras.apply(lambda row: func_escala(row), axis=1)
```

```python
# MONEDA

dict_moneda = {'Dólares':'$', 'Euro':'€'}
df_obras = df_obras.replace({'moneda':dict_moneda})
```

```python
# USOS LIMPIAR

df_obras.usos  = df_obras.usos.map(lambda x: x.lstrip('[]').rstrip('[]'))
df_obras.usos  = df_obras.usos.str.replace('"','')
df_obras.usos  = df_obras.usos.str.replace(',',', ')
```

```python
ind_impacto = ['im_sociedad','im_econom','im_ma','im_trans']
for i in ind_impacto:
    df_obras[i]  = df_obras[i].map(lambda x: x.lstrip('[]').rstrip('[]'))
    df_obras[i]  = df_obras[i].str.replace('"','')
    df_obras[i]  = df_obras[i].str.replace(',',', ')
```

```python
dict_ambitos = {'Argentina':'Argentina',
'Bolivia':'Bolivia',
'Brasil':'Brasil',
'Chile':'Chile',
'Colombia':'Colombia',
'Costa_Rica':'Costa Rica',
'Cuba':'Cuba',
'Ecuador':'Ecuador',
'El_salvador':'El Salvador',
'Espana':'España',
'Guatemala':'Guatemala',
'Honduras':'Honduras',
'Mexico':'México',
'Nicaragua':'Nicaragua',
'Panama':'Panamá',
'Paraguay':'Paraguay',
'Peru':'Perú',
'Portugal':'Portugal',
'Puerto_rico':'Puerto Rico',
'Republica_dominicana':'República Dominicana',
'Uruguay':'Uruguay',
'Venezuela':'Venezuela'}
```

```python
# DEFINIR ÁMBITOS

#dict_ambitos = dict(zip(df_places['País'], df_places['Ámbito']))
df_obras.loc[df_obras['pais'].isin(dict_ambitos.keys()), 'ambito'] = df_obras['pais'].map(dict_ambitos)
```

```python
df_obras = df_obras[['tipo_obra', 'escala', 'escala_tramos', 'usos', 'im_sociedad', 'im_econom', 'im_ma','im_trans', 'titulo_Obra', 'acr', 'año', 'direccion', 'pais', 'ambito', 'presupuesto', 'moneda','autores', 'img_url', 'ficha_url']]
```
